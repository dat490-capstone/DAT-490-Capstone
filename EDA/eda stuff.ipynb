```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\19258\Downloads\Medicare_FFS_2018_filtered.csv")

df["disease"] = df["category"]
df = df[(df["disease"] != "Any eye condition") & (df["disease"] != "Eye Exams")].copy()

df_race = df[
    (df["raceethnicity"] != "All races") &
    (df["sex"] == "Both sexes") &
    (df["age"] == "All ages")
].copy()

race_prevalence = (
    df_race.groupby(["raceethnicity", "disease"])["data_value"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(12, 6))
sns.barplot(
    data=race_prevalence,
    x="raceethnicity",
    y="data_value",
    hue="disease",
    palette="Set2"
)
plt.title("Eye Disease Prevalence by Race/Ethnicity (2018)", 
          fontsize=16, fontweight='bold')
plt.xlabel("Race/Ethnicity", fontsize=12)
plt.ylabel("Prevalence (%)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title="Eye Disease", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig(r"C:\Users\19258\Downloads\eye_disease_by_race.png", dpi=300, bbox_inches='tight')
plt.show()

df_sex = df[
    (df["raceethnicity"] == "All races") &
    (df["sex"] != "Both sexes") &
    (df["age"] == "All ages")
].copy()

sex_prevalence = (
    df_sex.groupby(["sex", "disease"])["data_value"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(10, 6))
sns.barplot(
    data=sex_prevalence,
    x="sex",
    y="data_value",
    hue="disease",
    palette="Set2"
)
plt.title("Eye Disease Prevalence by Sex (2018)", 
          fontsize=16, fontweight='bold')
plt.xlabel("Sex", fontsize=12)
plt.ylabel("Prevalence (%)", fontsize=12)
plt.legend(title="Eye Disease", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig(r"C:\Users\19258\Downloads\eye_disease_by_sex.png", dpi=300, bbox_inches='tight')
plt.show()

df_age = df[
    (df["raceethnicity"] == "All races") &
    (df["sex"] == "Both sexes") &
    (df["age"] != "All ages") &
    (df["age"] != "65 years and older") &
    (df["age"] != "18 years and older") &
    (df["age"] != "40 years and older")
].copy()

age_prevalence = (
    df_age.groupby(["age", "disease"])["data_value"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(14, 6))
sns.barplot(
    data=age_prevalence,
    x="age",
    y="data_value",
    hue="disease",
    palette="Set2"
)
plt.title("Eye Disease Prevalence by Age Group (2018)", 
          fontsize=16, fontweight='bold')
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Prevalence (%)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title="Eye Disease", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig(r"C:\Users\19258\Downloads\eye_disease_by_age.png", dpi=300, bbox_inches='tight')
plt.show()

df_state = df[
    (df["raceethnicity"] == "All races") &
    (df["sex"] == "Both sexes") &
    (df["age"] == "All ages")
].copy()

state_prevalence = (
    df_state.groupby(["locationabbr", "disease"])["data_value"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(16, 6))
sns.barplot(
    data=state_prevalence,
    x="locationabbr",
    y="data_value",
    hue="disease",
    palette="Set2"
)
plt.title("Eye Disease Prevalence by State (2018)", 
          fontsize=16, fontweight='bold')
plt.xlabel("State", fontsize=12)
plt.ylabel("Prevalence (%)", fontsize=12)
plt.xticks(rotation=90)
plt.legend(title="Eye Disease", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
