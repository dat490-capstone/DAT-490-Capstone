```python
import polars as pl
import pandas as pd
from pathlib import Path
```

```python
# Path to your raw CSVs
# === Paths to your raw CSVs ===
data_dir = Path("/Users/anviththumma/Desktop/School/DAT 490/Unfiltered Data/")

csv_files = [
    data_dir / "Medicare_FFS_2020.csv",
    data_dir / "Medicare_FFS_2021.csv",
    data_dir / "Medicare_FFS_2022.csv",
    data_dir / "Medicare_FFS_2018.csv",
    data_dir / "Medicare_FFS_2019.csv",
    data_dir / "Medicare_FFS_2017.csv",
    data_dir / "Medicare_FFS_2016.csv",
    data_dir / "Medicare_FFS_2015.csv",
    data_dir / "Medicare_FFS_2014.csv"
]
```

```python
# === Columns to keep (based on your analysis needs) ===
cols_to_keep = [
    "yearstart", "yearend",
    "locationabbr", "locationdesc", "geographiclevel",
    "topic", "category", "question", "response",
    "age", "sex", "raceethnicity", "riskfactor", "riskfactorresponse",
    "data_value", "data_value_type", "data_value_unit",
    "low_confidence_limit", "high_confidence_limit",
    "numerator", "sample_size"
]
```

```python
# === Load each CSV lazily and select only the desired columns ===
lazy_frames = []
for f in csv_files:
    lf = pl.scan_csv(f)
    # Correct way to get schema names
    schema_cols = lf.collect_schema().names()  # <-- add parentheses
    existing_cols = [c for c in cols_to_keep if c in schema_cols]
    lf = lf.select(existing_cols)
    lazy_frames.append(lf)
```

```python
# Directory to save new CSVs
output_dir = Path("/Users/anviththumma/Desktop/School/DAT 490/filtered_years/")
output_dir.mkdir(exist_ok=True)

# File names
output_files = [
    output_dir / "Medicare_FFS_2020_filtered.csv",
    output_dir / "Medicare_FFS_2021_filtered.csv",
    output_dir / "Medicare_FFS_2022_filtered.csv",
    output_dir / "Medicare_FFS_2018_filtered.csv",
    output_dir / "Medicare_FFS_2019_filtered.csv",
    output_dir / "Medicare_FFS_2017_filtered.csv",
    output_dir / "Medicare_FFS_2016_filtered.csv",
    output_dir / "Medicare_FFS_2015_filtered.csv",
    output_dir / "Medicare_FFS_2014_filtered.csv"
]

# Write each lazy frame
for lf, out_file in zip(lazy_frames, output_files):
    lf.collect().write_csv(out_file)
    print(f"Saved {out_file}")
```
